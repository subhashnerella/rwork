```python
import numpy
import math
from matplotlib import pyplot
```

```python
xc=0.75
yc=0.75
```

```python
R = 5+xc
a= xc-R
g=1-(xc/a)
cc=yc/g
Rc= R/g
theta = numpy.linspace(0,2*numpy.pi,150)
radius_matrix, theta_matrix = numpy.meshgrid(R,theta)
```

```python
X = radius_matrix * numpy.cos(theta_matrix)+xc
Y = radius_matrix * numpy.sin(theta_matrix)+yc
```

```python
phi= numpy.linspace(numpy.arctan(cc/a),numpy.pi-numpy.arctan(cc/a),150)
crad_matrix, cphi_matrix = numpy.meshgrid(Rc, phi)

XC=crad_matrix * numpy.cos(cphi_matrix)
YC=crad_matrix * numpy.sin(cphi_matrix)+cc
```

```python
%matplotlib inline

# plots the grid of points
size = 5
pyplot.figure(figsize=(size, size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.grid(True)
pyplot.plot(X, Y, color='r')
pyplot.plot(XC, YC, color='b')
pyplot.scatter(xc,yc)
pyplot.scatter(0,cc, color='b')
pyplot.plot([0.75,0,-4.951],[0.75,cc,0],color='#CD2305')
```

```python
E = X+1j*Y
EC = XC+1j*YC
```

```python
Z = E+((R-xc)**2)/E
ZC = EC+((R-xc)**2)/EC
```

```python
Xe ,Ye=Z.real ,Z.imag
Xce ,Yce=ZC.real ,ZC.imag
```

```python
%matplotlib inline

# plots the grid of points
size = 5
pyplot.figure(figsize=(size, size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.grid(True)
pyplot.ylim(-2*R,2*R)
#pyplot.xlim(x_start, x_end)
#pyplot.ylim(y_start, y_end)
pyplot.plot(Xe, Ye, color='#CD2305')
pyplot.plot(Xce, Yce, color='b')
```

```python
camber=-8*a*numpy.arctan(cc/a)/numpy.sin(2*numpy.arctan(cc/a))
re=40000
thE=2*camber/re**0.5
thR=2*thE
```

```python
gamma= numpy.linspace(1,g,41)
rgamma=numpy.linspace(1,g,41)*(a**2+cc**2)**0.5
rgamma1=numpy.linspace(1,g,41)*((a**2+cc**2)**0.5+thE)
rgamma2=numpy.linspace(1,g,41)*((a**2+cc**2)**0.5+thR)
gamma_mat, theta_mat=numpy.meshgrid(rgamma,theta)
gamma_mat1, theta_mat=numpy.meshgrid(rgamma1,theta)
gamma_mat2, theta_mat=numpy.meshgrid(rgamma2,theta)
X = gamma_mat * numpy.cos(theta_mat)+(1-gamma)*a
Y = gamma_mat * numpy.sin(theta_mat)+gamma*cc
X1 = gamma_mat1 * numpy.cos(theta_mat)+(1-gamma)*a
Y1 = gamma_mat1 * numpy.sin(theta_mat)+gamma*cc
X2 = gamma_mat2 * numpy.cos(theta_mat)+(1-gamma)*a
Y2= gamma_mat2 * numpy.sin(theta_mat)+gamma*cc
```

```python
# plots the grid of points
size = 5
pyplot.figure(figsize=(size, size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.grid(True)
pyplot.plot(X, Y, color='b');
#pyplot.scatter(X1, Y1, color='r')
#pyplot.plot(X2, Y2, color='y')
```

```python
zeta=X+1J*Y
Z=zeta+a**2/zeta
Xe ,Ye=Z.real ,Z.imag
```

```python
# plots the grid of points
size = 5
#pyplot.figure(figsize=(size, size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.grid(True)
pyplot.ylim(-2*R,2*R)
pyplot.plot(Xe, Ye, color='b');
#pyplot.scatter(X1, Y1, color='r')
#pyplot.plot(X2, Y2, color='y')
```

```python
# plots the grid of points
size = 5
pyplot.figure(figsize=(size, size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.grid(True)
pyplot.plot(X, Y, color='b')
pyplot.plot(X1, Y1, color='r')
pyplot.plot(X2, Y2, color='y')
```
